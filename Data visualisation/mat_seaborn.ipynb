## Univariate analysis -- Single variable

```python
data = pd.read_csv("Iris.csv")
```

```python
data.head()
```

```python
data['Species'].unique()
```

```python
cat_counts = data['Species'].value_counts()

cat_counts
```

**Bar Chart**

```python
x = cat_counts.index

x
```

```python
y = cat_counts.values

y
```

```python
plt.bar(x, y)

plt.figure(figsize = (12,8))

plt.bar(x, y)
```

```python
#plt.figure(figsize = (8,6))
plt.xticks(rotation = 90, fontsize = 15)
plt.xlabel('Species', fontsize = 25)
plt.ylabel('Number of species', fontsize = 25)
plt.title('No of species vs species')
plt.bar(x, y, width = 0.75, color = 'green')
plt.show()
```

```python
sns.countplot(x = 'Species', data = data, order = data['Species'].value_counts().index, color = 'orange')
```
